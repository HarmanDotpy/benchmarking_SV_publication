```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("../Data/cpu.ram.stats/cpu_mouse.csv") 
```

```python
data
```

```python
# gasv takes the longest, the two samples show similar behavior
import matplotlib.pyplot as plt
import seaborn as sns
g=sns.set_style("white")
g=sns.set_context("talk")
g = sns.catplot(x="sample_name", y="time",data=data,hue='Tool',aspect=1.8,kind='bar',capsize=.1,errwidth=3)

g.set(xlabel='Sample', ylabel='CPU-time (seconds)',yscale="log")
g=sns.despine()
plt.savefig('FigS20_time_sample.png')
plt.show(g)
```

```python
# gasv takes the most memory, the two samples show similar behavior
import matplotlib.pyplot as plt
import seaborn as sns
g=sns.set_style("white")
g=sns.set_context("talk")
g = sns.catplot(x="sample_name", y="memory (GB)",data=data,hue='Tool',aspect=1.8,kind='bar',capsize=.1,errwidth=3)

g.set(xlabel='Sample', ylabel='Memory (GB)',yscale="log")
g=sns.despine()
plt.savefig('FigS20_memory_sample.png')
plt.show(g)
```

```python
#Time taken by 1x is negligible as compared to full coverage sample 
g=sns.set_style("white")
g=sns.set_context("talk")
g = sns.catplot(x="Tool", y="time",data=data,hue='coverage',aspect=2,kind='bar',capsize=.1,errwidth=3)

g.set(xlabel='Tools', ylabel='CPU-time (seconds)',yscale="log")
g=sns.despine()
plt.savefig('FigS20_time_coverage.png')
plt.show(g)
```

```python
#Memory for full coverage and 1x is comparable 
g=sns.set_style("white")
g=sns.set_context("talk")
g = sns.catplot(x="Tool", y="memory (GB)",data=data,hue='coverage',aspect=2,kind='bar',capsize=.1,errwidth=3)
g.set(xlabel='Tools', ylabel='Memory (GB)',yscale="log")
g=sns.despine()
plt.savefig('FigS20_memory_coverage.png')
plt.show(g)
```

```python
#only contains coverage from 1x 
data_1x=data.loc[data['coverage'] == "1x"]
#only contains coverage from 64x 
data_64x=data.loc[data['coverage'] == "full_coverage"]
```

```python
data_1x
```

```python
#Variation of time for 1x with tool
g=sns.set_style("white")
g=sns.set_context("talk")
g = sns.catplot(x="Tool", y="time",data=data_1x,aspect=1.8,kind='bar',capsize=.1,errwidth=3)

g.set(xlabel='Tools (1x Coverage)', ylabel='CPU-time (seconds)',yscale="log")
g=sns.despine()
plt.savefig('FigS20_runtime_tools_1x.png')
plt.show(g)
```

```python
#Variation of memory for 1x with tool
g=sns.set_style("white")
g=sns.set_context("talk")
g = sns.catplot(x="Tool", y="memory (GB)",data=data_1x,aspect=1.8,kind='bar',capsize=.1,errwidth=3)

g.set(xlabel='Tools (1x Coverage)', ylabel='Memory (GB)',yscale="log")
g=sns.despine()
plt.savefig('FigS20_memory_tools_1x.png')
plt.show(g)
```
