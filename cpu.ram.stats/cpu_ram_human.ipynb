```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("cpu_human.csv") 
```

```python
data
```

```python
#clever has an outlier
import matplotlib.pyplot as plt
import seaborn as sns
g=sns.set_style("white")
g=sns.set_context("talk")
g = sns.catplot(x="chromosome", y="time",data=data,hue='Tool',aspect=1.8,kind='bar')
g.set(xlabel='Tools', ylabel='Run time (mins)')
g=sns.despine()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
g=sns.set_style("white")
g=sns.set_context("talk")
g = sns.catplot(x="chromosome", y="memory",data=data,hue='Tool',aspect=1.8,kind='bar')
g.set(xlabel='Tools', ylabel='Memory (MB)')
g=sns.despine()
```

```python
#Clever has an outlier 
g=sns.set_style("white")
g=sns.set_context("talk")
g = sns.catplot(x="Tool", y="time",data=data,hue='coverage',aspect=2.5,kind='bar')

g.set(xlabel='Tools', ylabel='Run time (mins)')
g=sns.despine()
plt.show(g)
```

```python
#Memory for full coverage and 1x is comparable 
g=sns.set_style("white")
g=sns.set_context("talk")
g = sns.catplot(x="Tool", y="memory",data=data,hue='coverage',aspect=2.5,kind='bar')

g.set(xlabel='Tools', ylabel='Memory (MB)')
g=sns.despine()
plt.show(g)
```
